```python
import pandas as pd
```

```python
user_df = pd.read_csv(r'C:\Users\WBrol\Downloads\BA test\Product_Data_Analyst_Test\Product_Data_Analyst_Test\user.csv',delimiter=";")
```

```python
user_df['Activity'] = "New user in system"
```

```python
user_df['created_utc'] = user_df['created']
```

```python
user_df.drop(columns=['created'], inplace=True)
```

```python
user_df
```

```python
bookings_df = pd.read_csv(r'C:\Users\WBrol\Downloads\BA test\Product_Data_Analyst_Test\Product_Data_Analyst_Test\bookings.csv',delimiter=";")
```

```python
bookings_df
```

```python
bookings_df_clean = bookings_df.loc[:, ['user_key', 'created_utc']]
```

```python
bookings_df_clean['Activity'] = "Create Booking"
```

```python
bookings_df_clean
```

```python
# Połączenie obu ramki danych
event_log = pd.concat([user_df, bookings_df_clean])

# Sortowanie po kolumnie 'user_key' i 'created_utc'
event_log.sort_values(by=['user_key', 'created_utc'], inplace=True)

# Resetowanie indeksu
event_log.reset_index(drop=True, inplace=True)

print(event_log)
```
